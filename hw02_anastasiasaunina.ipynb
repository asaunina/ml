#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.

```python
# устанавливаем библиотеку wget, чтобы загрузить файлы
!pip install wget
```

```python
# импортируем wget
import wget

# скачиваем зип-файл по ссылке
url = 'https://github.com/hse-ds/iad-intro-ds/raw/master/2023/homeworks/Data.zip'
filename = wget.download(url)
```

```python
# импортируем библиотеку zipfile, чтобы разархивировать скачанный зип-файл
import zipfile

# разархивируем все файлы
with zipfile.ZipFile(filename, 'r') as zip_ref:
    zip_ref.extractall()
```

```python
# импортируем библиотеку os, чтобы открыть все файлы в директории
import os

# все файлы были разархивированы в директории, у которой название такое же,
# как у зип-файла, но без .zip
path = filename.replace('.zip', '')

# в этих двух списках будут хранится таблички пандас, потом соединим их
students_info = []
students_marks = []
# для каждого файла в директории (в папке)
for fi in sorted(os.listdir(path)):
    # если в названии файла есть 'Students_info_'
    if 'Students_info_' in fi:
        # читаем csv файл, делаем его таблицей в пандасе и добавляем в список
        students_info_i = pd.read_csv(f'{path}/{fi}')
        students_info.append(students_info_i)
    # аналогично, если в названии есть 'Students_marks_'
    elif 'Students_marks_' in fi:
        # делаем файл таблицей в пандасе и добавляем в список
        students_marks_i = pd.read_csv(f'{path}/{fi}')
        students_marks.append(students_marks_i)
# создаем общую таблицу всех студентов
info = pd.concat(students_info)
# создаем общую таблицу всех оценок
marks = pd.concat(students_marks)
# соединяем получившиеся таблицы по индексу, outer join'ом
# (чтобы нулевые значения тоже сохранились)
df = pd.merge(info, marks, on="index", how="outer")
# выводим таблицу (5 произвольных строк)
df.sample(5)
```

**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

Для каждой этнической группы выведите средний балл за экзамен по чтению

```python
# чтобы засечь время
%%time
# создаем словарь с национальностями, куда будем записывать баллы по чтению
ethnicities = {}
# для каждой строчки в таблцие
for i in range(len(df['race/ethnicity'])):
    # если национальность еще не было в словаре
    if df['race/ethnicity'][i] not in ethnicities:
        # создаем для нее пустой словарь
        ethnicities[df['race/ethnicity'][i]] = []
        # затем добавляем в список балл по чтению для каждой строки
        ethnicities[df['race/ethnicity'][i]].append(df['reading score'][i])
    # если национальность была в словаре
    else:
        # добавляем в список балл по чтению
        ethnicities[df['race/ethnicity'][i]].append(df['reading score'][i])
# для каждого ключа в словаре (каждой национальности)
for key in ethnicities:
    # присваиваем вместо списка среднее его значение
    ethnicities[key] = sum(ethnicities[key]) / len(ethnicities[key])
# сортируем словарь по алфавиту, чтобы было красивее
ethnicities = dict(sorted(ethnicities.items()))
ethnicities
```

Время выполнения больше почти в 10 раз (3.24 и 31 ms)

Для каждого уровня образования выведите минимальный балл за экзамен по письму

```python
# чтобы засечь время
%%time
# создаем словарь с уровнями образования, куда будем записывать баллы по письму
educations = {}
# для каждой строчки в таблице
for i in range(len(df['education'])):
    # если уровень образования еще не встречался в словаре
    if df['education'][i] not in educations:
        # создаем для этого уровня образования словарь, где будут баллы по письму
        educations[df['education'][i]] = []
        # добавляем балл по письму
        educations[df['education'][i]].append(df['writing score'][i])
    # если уровень образования уже был в словаре
    else:
        # добавляем балл по письму
        educations[df['education'][i]].append(df['writing score'][i])
# для каждого уровня образования в словаре
for key in educations:
    # вместо списка баллов оставляем минимальный
    educations[key] = min(educations[key])
educations = dict(sorted(educations.items()))
educations
```

##### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
# функция перевода балла в оценку (граничный балл относится к оценке выше)
def GetGrade(average_mark):
    if average_mark >= 90:
        return 'A'
    elif average_mark >= 80:
        return 'B'
    elif average_mark >= 70:
        return 'C'
    elif average_mark >= 60:
        return 'D'
    elif average_mark >= 50:
        return 'E'
    else:
        return 'F'

# создаем новый столбец, применяя функцию GetGrade к каждому элементу, который определяется как
# среднее по трем полученным баллам за экзамены
df['Grade'] = np.array(list(map(GetGrade,
 ((df['math score'] + df['reading score'] + df['writing score']) / 3))))
df
```
